```python
from sklearn import datasets
import numpy as np
import random
import math
import matplotlib.pyplot as plt
 
iris = datasets.load_iris()
X = iris.data[:, :2]
y = (iris.target != 0) * 1
 
 
def shuffle(x, y):
    random.seed(1)
    for s in range(len(x)):
        p = []
        R = random.random()
        if R <= 0.7:
            a = x[s]
            p = [a[0], a[1], y[s]]
            train_set.append(p)
        elif 0.7 <= R < 0.85:
            a = x[s]
            p = [a[0], a[1], y[s]]
            val_set.append(p)
        else:
            a = x[s]
            p = [a[0], a[1], y[s]]
            test_set.append(p)
 
 
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
 
 
def logistic_regression(lr):
 
    Q = [random.random(), random.random(), random.random()]
 
    for t in train_set:
        concatet.append([1, t[0], t[1]])
        Y.append(t[2])
    #print(Y)
    for i in range(0, 1000): 
        tj = 0
 
        for t in range(len(concatet) - 1):
            X = concatet[t]
            z = np.dot(X, Q)
            h = sigmoid(z)
            y = Y[t]
 
            j = -y * np.log(h) - (1 - y) * np.log(1 - h)
            tj = tj + j
 
            dv = np.array(X) * (h - y)
 
            Q = np.array(Q) - dv * lr
 
            # dv=[x*(h-y) for x in X]
            # dv2=[x*lr for x in dv]
            # Q=np.array(Q)-np.array(dv2)
 
        train_loss.append(tj / len(concatet))
        
        epoch.append(i)
    return Q
 
 
def logistic_regression_graph():
 
    plt.plot(epoch, train_loss)
    plt.xlabel("epoch")
    plt.ylabel("train loss")
    plt.show()
 
 
def val_acc(val_set):
    correct = 0
    for t in val_set:
        concatet.append([1, t[0], t[1]])
 
    for s in range(len(val_set)):
        X = concatet[s]
        z = np.dot(X, Q)
 
        h = sigmoid(z)
 
        if h >= 0.5:
            h = 1
        else:
            h = 0
 
        v = val_set[s]
 
        if h == v[2]:
            correct += 1
 
    val_acc = (correct / len(val_set)) * 100
    return val_acc
 
 
if __name__ == '__main__':
 
    lr = [0.1, 0.01, 0.001, 0.0001,0.00001]
    for i in lr:
        train_set = []
        val_set = []
        test_set = []
 
        shuffle(X, y)
 
        Q = []
 
        concatet = []
        Y = []
        train_loss = []
        epoch = []
        Q = logistic_regression(i)
        concatet = []
        print("For lr =",i,"accuracy is ",val_acc(val_set))
 
        logistic_regression_graph()
```
